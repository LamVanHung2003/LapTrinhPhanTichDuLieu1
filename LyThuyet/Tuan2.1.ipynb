```python
# Import thư viện
import pandas as pd
import seaborn as sns
# Đưa dữ liệu vào datafame
df = pd.read_csv('E:\LTPhanTichDuLieu/dulieuxettuyendaihoc.csv', header=0, encoding='utf-8', delimiter=',')
```

```python
# Cắt lấy cái cột cần sử dụng

df = df[['T5','T6','GT','DT','KV','KT','NGONNGU','TOANLOGICPHANTICH','GIAIQUYETVANDE',
         'NGAYTHI','DINHHUONGNGHENGHIEP']]

df
```

```python
df.head(5)
```

```python
#Xoá bỏ các dòng dũ liệu rỗng
df.dropna(how='all', inplace=True)
```

```python
# Xoá bỏ các dòng bị trùng
df.drop_duplicates(inplace=True)
```

```python
# Điền giá trị thiếu
df['DT'].fillna('KINH', inplace=True)
# Lưu ý: với biến định tính ta có thể thay
# bằng giá trị yếu vị (mode)
#df['DT'].fillna(df['DT'].mode()[0], inplace=True)
```

```python
# Điền thiếu giá trị phần NGONNGU bằng 0 (nếu có)
df['NGONNGU'].fillna(0,inplace=True)
# Điền thiếu giá trị phần LOGIC bằng trung bình (nếu có)
df['LOGIC'].fillna(df['LOGIC'].mean(), inplace=True)
# Điền thiếu giá trị phần UNGXU bằng trung vị (nếu có)
df['UNGXU'].fillna(df['UNGXU'].median(), inplace=True)
# Lưu ý: Với biến định lượng thì ta nên thay thê bằng trung vị

# Sinh viên tự nghiên cứu:
# Điền thiếu T6 bằng phương pháp interpolate (nội suy)
# Nội suy từ những cái đã có => cái đang có
# Dựa vào cái giá trị đã có để suy diễn ra dữ liệu bị thiếu(Kinh doanh tháng 6 bị thiếu
# => Dựa vào các giá trị tháng khác=> tính trung bình, trung vị gì đó để => 6) soward, backwward
# Lưu ý: Interpolate dùng cho dữ liệu có tính thứ tự (tuần tự)
# 
```

```python
# Định tính: Scalan, AI, Mode
# Định lượng: Scalan, Mean, Median, X/s(Xác suất), AI
```

```python
# Phần 2: Kỹ thuật Featur Enginerring
# (Thường dùng cho Machine Learning)
# Nếu chỉ là xử lý phân tích dữ liệu thì ta gọi là New Attribute

# Đây là kỹ thuật tạo thêm hoặc biến đổi số liệu sẵn có thành các
# biến số mới phù hợp nghiệp vụ để phân tích

# Vd: Tạo 1 cột mới từ điểm cột theo quy định của Bộ GD thì đó là đưa nghiệp vụ vào tạo thành 1 table có ý nghĩa
 # Tạo biến TBTOAN: Trung bình toán lớp 12
 
df['TBTOAN'] = (df['T5']+df['T6'])/2
df
```

```python
# Tạo biến XEPLOAI: Đánh giá môn toán dựa trên df['TBTOAN]
df.loc[df['TBTOAN'] < 5.0, 'XEPLOAI'] = 'FAIL'

df.loc[(df['TBTOAN'] >= 5.0) & (df['TBTOAN'] < 7.0), 'XEPLOAI'] = 'FAIR'

df.loc[(df['TBTOAN'] >= 7.0) & (df['TBTOAN'] < 9.0), 'XEPLOAI'] = 'GOOD'

df.loc[df['TBTOAN'] >= 9.0 , 'XEPLOAI'] = 'EXCEL'
```

```python
df[['TBTOAN','XEPLOAI']].head(5)
```

```python
# Tạo biến nhóm khối thi: NHOMKT thoả mãn
# A1: G1
# C: G3
# D1: G3
# A: G1
# B: G2
# ...

dict_map = {
    'A1':'G1','C':'G3','D1':'G3','A':'G1','B':'G2'
}

df['NHOMKT'] = df['KT'].map(dict_map)
```

```python
# Tạo biến số điểm cộng: DIEMCONG
# ...
# Nếu NHOMKT thuộc nhóm G1, G2 và TBTOAN >= 5.0 thì là 1.0
# Ngược lại thì 0.0
# ...
# Dùng tạo ra các cột có điều kiện phức tạp

def fplus(x,y):
    if (x == 'G1' or x == 'G2') and (y >= 5.0):
        return 1.0
    else:
        return 0.0

df['DIEMCONG'] = list(map(fplus,df['NHOMKT'],df['TBTOAN']))

df[['NHOMKT','TBTOAN','DIEMCONG']].head(5)
```

```python
import numpy as np
# Khi biến định tính dùng làm nhóm tổng hợp
# Có nhiều hơn 2 giá trị hoặc có giá trị thiếu của
# Điểm NGONNGU trên từng nhóm
# Thì ta cần dùng hàm tổng hợp thông qua thư viện numpy
sns.barplot(x='GT', y='NGONNGU', hue='KT', data=df, estimator=np.max)
plt.show()
```

```python
# ...
# Sinh viên lưu ý:
# - Với biến định tính thì cta chỉ có một hàm tổng là
# hàm COUNT, MODE
# - Với định lượng thì ta có thể sử dụng các hàm tổng hợp như:
# COUNT, MAX, MIN, MEAN, MEDIAN, MODE, SUM, STD
# ...
```

```python
# ...
# Biểu đồ PIE
# Mục đích: Trực quan dữ liệu theo nhóm KT theo tỉ lệ phần trăm
# ...

gb = df.groupby(['KT'])['KT'].agg(['count'])

labels = gb.index # Labels = ['A','A1','B','C','D1']
data = list(gb['count']) # data = [49,6,9,14,22]

colors = sns.color_palette('pastel') # Tạo bảng màu

plt.pie(data, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True)

plt.show()
```

```python
# ...
# Sinh viên tự viết thống kê tỉ lệ phần trăm cho KV
# ...
gb = df.groupby(['KV'])['KV'].agg(['count'])

labels = gb.index
data = list(gb['count'])

colors = sns.color_palette('pastel')

plt.pie(data, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True)

plt.show()
```

```python
# ...
# Trực quan tỉ lệ phần trăm tổng điểm DIEMCONG trên từng nhóm KV
# ...

gb = df.groupby(['KV'])['DIEMCONG'].agg(['sum'])

labels = gb.index

data = list(gb['sum'])

colors = sns.color_palette('pastel')

plt.pie(data, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True)

plt.show()
```
