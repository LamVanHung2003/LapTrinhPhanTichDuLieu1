```python
import pandas as pd
df=pd.read_csv('E:\LTPhanTichDuLieu\dulieuxettuyendaihoc.csv',header=0,delimiter=',',encoding='UTF-8')
```

```python
import seaborn as sns
```

```python
# Cắt lấy cái cột cần sử dụng

df = df[['T5','T6','GT','DT','KV','KT','NGONNGU','TOANLOGICPHANTICH','GIAIQUYETVANDE',
         'NGAYTHI','DINHHUONGNGHENGHIEP']]

df
```

```python
df.info()
```

```python
df.rename(columns={'TOANLOGICPHANTICH':'LOGIC','GIAIQUYETVANDE':'UNGXU','DINHHUONGNGHENGHIEP':'HUONGNGHIEP'},inplace=True)
```

```python
df.head(5)
```

```python
#xoa cac du lieu rong 
df.dropna(how='all',inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.isna().transpose(),cmap='YlGnBu',cbar_kws={'label':'DLTHIEU'})
plt.savefig('missingdata.png',dpi=100)
plt.show()
```

```python
df['DT'].fillna('Kinh',inplace=True)
```

```python
df['NGONNGU'].fillna(0,inplace=True)
```

```python
df['LOGIC'].fillna(df['LOGIC'].mean(),inplace=True)
```

```python
df['UNGXU'].fillna(df['UNGXU'].median(),inplace=True)
```

```python
df['TBTOAN']=(df['T5']+df['T6'])/2
df
```

```python
df.loc[df['TBTOAN']<5.0,'XEPLOAI']='FAIl'
df.loc[(df['TBTOAN']>=5)&(df['TBTOAN']<7.0),'XEPLOAI']='FAIR'
df.loc[(df['TBTOAN']>=7.0)&(df['TBTOAN']<9),'XEPLOAI']='GOOD'
df.loc[df['TBTOAN']>=9.0,'XEPLOAI']='EXCEL'
df
```

```python
df.loc[df['XEPLOAI']=='GOOD'].count()
```

```python
dict_map={
    'A1':'G1','C':'G3','D1':'G3','A':'G1','B':'G2',
}
df['NHOMKT']=df['KT'].map(dict_map)
df
```

```python

def fplus(x,y):
    if (x == 'G1' or x == 'G2') and (y >= 5.0):
        return 1.0
    else:
        return 0.0

df['DIEMCONG'] = list(map(fplus,df['NHOMKT'],df['TBTOAN']))

df[['NHOMKT','TBTOAN','DIEMCONG']].head(5)
```

```python
gb=df.groupby(['KV'])['DIEMCONG'].agg(['sum'])

labels=gb.index
data=list(gb['sum'])

colors=sns.color_palette('pastel')

plt.pie(data,labels=labels,colors=colors,autopct='%1.1f%%',shadow=True)
plt.show()
```
