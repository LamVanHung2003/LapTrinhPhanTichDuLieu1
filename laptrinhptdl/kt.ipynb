```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
df =pd.read_csv('sales_data_sample.csv',encoding = 'unicode_escape')
df.head()
```

```python
# Nhóm dữ liệu theo quốc gia và tính tổng doanh số bán hàng
country_sales = df.groupby('COUNTRY')['SALES'].sum().reset_index()

# Sắp xếp theo doanh số bán hàng giảm dần
sorted_sales = country_sales.sort_values('SALES', ascending=False)

# Lấy quốc gia ở vị trí thứ 5
if len(sorted_sales) >= 5:
    fifth_country = sorted_sales.iloc[4]['COUNTRY']
    fifth_sales = sorted_sales.iloc[4]['SALES']
    print("Quốc gia ở vị trí thứ 5:", fifth_country)
    print("Tổng doanh số bán hàng ở quốc gia ở vị trí thứ 5:", fifth_sales)
else:
    print("Không đủ dữ liệu để xác định quốc gia ở vị trí thứ 5.")
import pandas as pd

# Assuming you have a DataFrame named 'data' containing the mentioned columns
# Sort the DataFrame by 'QUANTITYORDERED' column in descending order
sorted_data = df.sort_values('QUANTITYORDERED', ascending=False)

# Get the country in the fifth row of the sorted DataFrame
country_row = sorted_data.iloc[4]

# Retrieve the name of the country
country_name = country_row['COUNTRY']

# Filter the DataFrame to include only rows with the selected country
filtered_data = df[df['COUNTRY'] == country_name]

# Calculate the sum of 'SALES' column for the filtered DataFrame
total_sales = filtered_data['SALES'].sum()

print("Country ranked fifth in terms of quantity:", country_name)
print("Total sales in the fifth-ranked country:", total_sales)

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named 'data' containing the mentioned columns
# Retrieve the relevant data from the dataset
order_sizes = df['DEALSIZE']
quantities = df['QUANTITYORDERED']

# Group the data by 'DEALSIZE' and calculate the total quantity for each order size
grouped_data = df.groupby('DEALSIZE')['QUANTITYORDERED'].sum()

# Calculate the percentage of each order size
percentages = grouped_data / quantities.sum() * 100

# Create a pie chart
plt.pie(percentages, labels=grouped_data.index, autopct='%1.1f%%')
plt.title("Percentage of Quantity by Order Size")
plt.axis('equal')

# Show the pie chart
plt.show()
```
