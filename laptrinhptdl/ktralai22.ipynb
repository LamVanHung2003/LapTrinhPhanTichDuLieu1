```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression

# Đường dẫn tới file "boston.csv"
file_path = "boston.csv"

# Đọc file CSV bằng pandas
data = pd.read_csv(file_path)

# Chuẩn bị dữ liệu đầu vào và đầu ra
X = data[['age', 'lstat']]
y = data['medv']

# Khởi tạo StandardScaler
scaler = StandardScaler()

# Chuẩn hóa dữ liệu đầu vào
X_scaled = scaler.fit_transform(X)

# Chia tập dữ liệu thành tập huấn luyện và tập kiểm tra
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.1, random_state=10)

# Xây dựng mô hình hồi quy tuyến tính
model = LinearRegression()

# Huấn luyện mô hình trên tập huấn luyện
model.fit(X_train, y_train)

# Dự báo giá trị medv
X_input = [[37.5, 4.65]]
X_input_scaled = scaler.transform(X_input)
prediction = model.predict(X_input_scaled)

# In giá trị dự báo medv
print("Giá trị dự báo medv:", prediction[0])
```
