```python
df = df[['T1','T2','T3','T4','T5','T6','GT','TOANLOGICPHANTICH']]
```

```python
df.rename(columns={'TOANLOGICPHANTICH':'LOGIC'}, inplace=True)
```

```python
# Tập dữ liệu Input (đặc trưng) và Output (mục tiêu)
X = df[['T1','T2','T3','T4','T5','T6','F','M']].values #input
Y = df[['LOGIC']].values #input

# Đảm bảo dữ liệu là kiểu số thực

X = X.astype(float)
Y = Y.astype(float)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2, random_state=16)
```

```python
def x_function(X):
    return X**2;
```

```python
from sklearn.preprocessing import FunctionTransformer
from sklearn.preprocessing import StandardScaler
import numpy as np

X_train = x_function(X_train)
# sc = StandardScaler()
# X_train = sc.fit_transform(X_train)
```

```python
X_train
```

```python
import matplotlib.pyplot as plt

X_train_df = pd.DataFrame(X_train)


tmp = X_train_df.iloc[:,0:6]
tmp.hist()

```

```python
from sklearn.preprocessing import MinMaxScaler
# Scale cho Input
sc = MinMaxScaler()
X_train = sc.fit_transform(X_train)
```

```python
X_train_df = pd.DataFrame(X_train)

tmp = X_train_df.iloc[:,0:6]
tmp.boxplot()
```

```python


from sklearn import linear_model

model = linear_model.LinearRegression()
model.fit(X_train,Y_train)


# To retrive the intrcept
print(model.intercept_)

# For retrieving the slope (coefficient):
print(model.coef_)
```

```python
# Đánh giá mô hình
import numpy as np
from sklearn import metrics

X_test = x_function(X_test)
X_test = sc.transform(X_test)

Y_test_pred = model.predict(X_test)
Y_train_pred = model.predict(X_train)

print('Score or R-Squared(test): ', model.score(X_test,Y_test))
print('Score or R-Squared(train): ', model.score(X_train,Y_train))
print('Mean Absolute Error(test):', metrics.mean_absolute_error(Y_test,Y_test_pred))
print('Mean Absolute Error(train):', metrics.mean_absolute_error(Y_train,Y_train_pred))
print('Mean Squared Error(test):', metrics.mean_squared_error(Y_test,Y_test_pred))
print('Mean Squared Error(train):', metrics.mean_squared_error(Y_train,Y_train_pred))
print('Root Mean Squared Error(test): ', np.sqrt(metrics.mean_squared_error(Y_test,Y_test_pred)))
print('Root Mean Squared Error(train): ', np.sqrt(metrics.mean_squared_error(Y_train,Y_train_pred)))
print('Mean Absolute Percentage Errortest: ', metrics.mean_absolute_percentage_error(Y_test,Y_test_pred))
print('Mean Absolute Percentage Errortrain: ', metrics.mean_absolute_percentage_error(Y_train,Y_train_pred))
```

```python
# Độ lệch validation giữa tập train và test không đáng kể, score khá thấp
```
