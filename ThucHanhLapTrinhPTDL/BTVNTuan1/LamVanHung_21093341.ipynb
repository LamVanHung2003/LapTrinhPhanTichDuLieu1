```python
import pandas as pd
thietbi= pd.read_csv('thietbi.csv')
nguoidung_thietbi=pd.read_csv('nguoidung_thietbi.csv')
nguoidung_sudung=pd.read_csv('nguoidung_sudung.csv')
```

```python
#3 đổi tên các cột dữ liệu
thietbi.rename(columns={'Retail Branding':'Branding','Marketing Name':'MarketingName'},inplace=True)
```

```python
thietbi.head()
```

```python
#4
dfKetQua=pd.merge(nguoidung_sudung,nguoidung_thietbi[['use_id','platform','device']],on='use_id')
```

```python
dfKetQua.head(5)
```

```python
#hoặc left-join
dfKetQua=pd.merge(nguoidung_sudung,nguoidung_thietbi[['use_id','platform','device']],on='use_id',how='left')
```

```python
#5 
#inner-join
dfKetQua=pd.merge(dfKetQua,thietbi[['Branding','Model']],left_on='device',right_on='Model')
```

```python
#left-join
dfKetQua=pd.merge(dfKetQua,thietbi[['Branding','Model']],left_on='device',right_on='Model',how='left')
```

```python
#6 hãy liệt kê 5 dòng đầu tiên các mẫu thiết bị
dfKetQua[dfKetQua.device.str.startswith('GT')].head(5)
```

```python
#7 thống kê trung bình số phút hàng tháng, trung bình sms và trung bình data,số lượng sử dụng của các nhãn hàng
dfKetQua.groupby('Branding').agg({
    'outgoing_mins_per_month':'mean',
    'outgoing_sms_per_month':'mean',
    'monthly_mb':'mean',
    'use_id':'count'
})
```

```python
#8 trực quan số lượng người dùng các nhãn hàng
import seaborn as sns
import matplotlib.pyplot as plt
sns.countplot(y='Branding',data=dfKetQua)
plt.show()
```

```python
# trực quan hóa dữ liệu trung bình phút gọi, trung bình sms và dung lượng sử dụng trên tưng nhóm nhãn hàng bán lé
import numpy as np
sns.barplot(y='Branding',x='outgoing_mins_per_month',data=dfKetQua,estimator=np.mean,ci=None,orient='horizontal')
plt.show()
```

```python
#10 khảo sát tương quan giữa phút gọi,sms và dung lượng trên từng nhóm nhãn hàng
sns.scatterplot(x='outgoing_mins_per_month',
               y='outgoing_sms_per_month',
               size='monthly_mb',
               alpha=0.5,
               hue='Branding',
               data=dfKetQua)
plt.show()
```
