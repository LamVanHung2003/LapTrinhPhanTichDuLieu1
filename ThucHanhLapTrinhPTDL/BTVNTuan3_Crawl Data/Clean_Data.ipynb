```python
import pandas as pd
```

```python
# a. Đọc dữ liệu lên data frame
df=pd.read_csv('inspirational_quotes.csv', header=0,delimiter=',', encoding='ISO 8859-1')
df.head()
```

```python
#b. Dữ liệu rỗng
df.dropna()
```

```python
# c. Dữ liệu trùng
df.drop_duplicates()
```

```python
# d. Dữ liệu sai định dạng
df.replace({r'[^\x00-\x7F]+':''},regex=True,inplace=True)
print(df)
```

```python
# e. Dữ liệu lỗi Unicode
df[['theme','img','url','lines']]=df[['theme','img','url','lines']].astype("string")
```

```python
# f. Dữ liệu chứa nhiều thông tin cần tách ra
# tách dữ liệu cột lines thành dòng riêng biệt
df['quote_lines']=df['lines'].str.split('\n')

df.head()
```

```python
# g. Dữ liệu thiếu
# kiểm tra dữ liệu thiếu
df.isna()
```

```python
#lưu trữ dữ liệu đã xử lý 
df.to_csv('inspirational_quotes_clean.csv')
```
