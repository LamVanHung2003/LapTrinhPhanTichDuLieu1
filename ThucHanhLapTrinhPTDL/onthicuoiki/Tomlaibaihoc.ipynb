```python
# Data->MinMaxScaler -> modal
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import linear_model
import numpy as np
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import FunctionTransformer
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import PolynomialFeatures
from sklearn import metrics 

```

```python
df= pd.read_csv('dulieuxettuyendaihoc.csv', header=0,delimiter=',', encoding ='utf-8')
df
```

```python
df.rename(columns={'TOANLOGICPHANTICH':'LOGIC','GIAIQUYETVANDE':'UNGXU','DINHHUONGNGHENGHIEP':'HUONGNGHIEP'},inplace=True)
```

```python
df = df[['T1','T2','T3','T4','T5','T6','GT','LOGIC']]
df
```

```python
df.describe()
```

```python
GT=pd.get_dummies(df['GT']).astype(int)
GT
# True = 1
# Flase =0
```

```python
df=pd.concat([df,GT],axis=1)
df.head(5)
df.drop('GT',axis=1,inplace=True)
df
```

```python
x=df[['T1','T2','T3','T4','T5','T6','F','M']].values # input frature dac Trung
y=df['LOGIC'].values # output muc tieu (Target)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size= 0.2,random_state=16)
```

```python
x_train
```

```python
# x*x
x_train_BP = np.square(x_train)
x_train_BP[0]
# hàm X_train như nào thì X_Test cũng phải i chang
```

```python
# Sqrt
x_train_SQRT = np.sqrt(x_train_BP)
x_train_SQRT[0]
```

```python
# # log
# x_train_log = np.log(x_train_SQRT)
# x_train_log
```

```python
# minmaxScaler 
scaler = MinMaxScaler()
x_train_scaled = scaler.fit_transform(x_train_SQRT)
x_train_scaled[0]
```

```python
# poly 
poly = PolynomialFeatures(degree=2)
x_train_poly = poly.fit_transform(x_train_scaled)
```

```python
# nào khúc cuối mới tạo regressor và gắn nó cho phần X_train 
regressor = LinearRegression()
regressor.fit(x_train_poly, y_train)
```

```python
# Hệ số tự do
print(regressor.intercept_)
# Hệ số gốc
print(regressor.coef_)
```

# X_Test gắn hàm số dô


```python
# x*x
x_test_BP = np.square(x_test)
# Sqrt
x_test_SQRT = np.sqrt(x_test_BP)
# # minmaxScaler 
scaler = MinMaxScaler()
x_test_scaled = scaler.fit_transform(x_test_SQRT)
# # poly 
# poly = PolynomialFeatures(degree=2)
# x_test_poly = poly.fit_transform(x_test_scaled)
```

```python
import numpy as np
from sklearn import metrics

scaler.fit(x_test_scaled)

#đánh giá mô hình
# Biến đổi dữ liệu kiểm thử
x_test_scaled = poly.transform(x_test_scaled)
y_test_pred = regressor.predict(x_test_scaled)

print('Score or R-Squared: ', regressor.score(x_test_scaled, y_test))

# print('Mean Absolute Error: ', metrics.mean_absolute_error(y_test, y_test_pred))
# print('Mean Squared Error:     ', metrics.mean_squared_error(y_test, y_test_pred))
print('Root Mean Squared Error: ', np.sqrt(metrics.mean_squared_error(y_test, y_test_pred)))
print('Mean Absolute Percentage Error: ', metrics.mean_absolute_percentage_error(y_test, y_test_pred))
```
