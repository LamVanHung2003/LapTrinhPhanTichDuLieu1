```python
import pandas as pd 
column_name =['ID','Name','Age','Weight','m006','m0612','m1218','f006','f0612','f1218']
df= pd.read_csv('patient_heart_rate.csv',names=column_name)
```

```python
df.head(10)
```

```python
df[['firstname','Lastname']]=df['Name'].str.split(expand=True)
df=df.drop('Name',axis=1)
df.head(5)
```

```python
weight = df['Weight']
for i in range (0,len(weight)):
    x=str(weight[i])
    if 'lbs' in x[-3:]:
        x=x[:-3:]
        float_x = float(x)
        y=int(float_x/2.2)
        #y=str(y)
        weight[i]=y
    if'kgs' in x[-3:]:
        x= x[:-3:]
        float_x=float(x)
        weight[i]=x
```

```python
df.head(5)
```

```python
df.rename(columns={'Weight':'Weight_kgs'},inplace=True)
df.head(10)
```

```python
#xóa các dòng dữ liệu rỗng

df=df.dropna()
df.head()
```

```python
#xóa các dữ liệu trùng
df=df.drop_duplicates(subset=['firstname','Lastname','Age','Weight_kgs'])
```

```python
df.head(10)
```

```python
df.firstname.replace({r'[^\x00-\x7F]+':''},regex=True,inplace=True)
df.Lastname.replace({r'[^\x00-\x7F]+':''},regex=True,inplace=True)
print(df)
```

```python
df['Age'].fillna(df['Age'].mode()[0],inplace=True)
```

```python
#Thay giá trị thiếu cân nặng bằng giá trị trung vị 
df['Weight_kgs'].fillna(df['Weight_kgs'].median(),inplace=True)
```

```python
df.head()
```

```python
# 11. phân rã các cột chứa nhiều thông tin 
# Tạo melt dữ liệu để có cột gender_time
df = pd.melt(df, id_vars=['Id', 'Age', 'Weight_kgs', 'Firstname', 'Lastname'], 
             value_name='PulseRate', var_name='gender_time').sort_values(['Id', 'Age', 'Weight_kgs', 'Firstname', 'Lastname'])
```

```python
# Tạo data frame tạm là kết quả của việc tách cột gender_time
df_temp = df['gender_time'].str.extract("(\D)(\d+)(\d{2})", expand=True)
```

```python
# Đặt tên cột cho data frame tạm
df_temp.columns = ['Gender', 'Lower_hour', 'Upper_hour']
```

```python
# Nối data frame tạm vào data frame ban đầu
df = pd.concat([df, df_temp], axis=1)
```

```python
# Bỏ qua cột gender_time
df = df.drop(['gender_time'], axis=1)
```

```python
#12. Loại bỏ hết các dòng dữ liệu thừa là những dòng có phần PulseRate có dấu -
import numpy as np

df = df.replace('-', np.nan).dropna(subset=['PulseRate'])
df.head(10)
```

```python
# 13. Nhận thấy có những bệnh nhân chưa ghi nhận họ tên (ví lý do nào đó)
df['Firstname'].isnull().sum()
df['Lastname'].isnull().sum()
# Thay họ tên thành Unknown
df['Firstname'].fillna('Unknown', inplace=True)
df['Lastname'].fillna('Unknown', inplace=True)
```

```python
# 14. Sau khi xử lý thì index của dòng dữ liệu đã thay đổi lung tung, ta cần reset index lại cho theo khuôn mẫu
df = df.reset_index()
```

```python
# 15. Sau đó, lưu trữ dữ liệu đã xử lý thành công với tên file patient_heart_rate_clean.csv
df.to_csv('patient_heart_rate_clean.csv')
```
